Q1: What is Simple Linear Regression?

A1: Simple Linear Regression is a statistical method used to model the relationship between two variables: one independent variable (X) and one dependent variable (Y). It assumes a linear relationship and represents it with the equation:
Y = mX + c,
where m is the slope and c is the intercept. The goal is to find the best-fitting straight line through the data points.



Q2: What are the key assumptions of Simple Linear Regression?

A2: The key assumptions are:

Linearity: The relationship between X and Y is linear.

Independence: Observations are independent of each other.

Homoscedasticity: Constant variance of errors.

Normality of errors: The residuals (errors) are normally distributed.

No or minimal multicollinearity: (Only applicable in multiple regression, but worth knowing).



Q3: What does the coefficient m represent in the equation Y = mX + c?

A3: The coefficient m represents the slope of the line. It indicates the change in the dependent variable (Y) for a one-unit change in the independent variable (X). In other words, it shows how much Y increases or decreases when X increases by 1.

Q4: What does the intercept c represent in the equation Y = mX + c?

A4: The intercept c is the value of Y when X is 0. It shows where the regression line crosses the Y-axis. It's the starting point of the line when the independent variable has a value of zero.

Q5: How do we calculate the slope m in Simple Linear Regression?

A5: The slope m is calculated using the formula:
m = Σ[(Xi - X̄)(Yi - Ȳ)] / Σ[(Xi - X̄)²]
Where:

Xi and Yi are the individual data points

X̄ is the mean of X values

Ȳ is the mean of Y values

Q6: What is the purpose of the least squares method in Simple Linear Regression?

A6: The least squares method aims to find the best-fitting line by minimizing the sum of the squares of the residuals (the differences between the actual Y values and the predicted Y values). This ensures the line is as close as possible to all data points.



Q7: How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

A7: R², or the coefficient of determination, measures the proportion of the variance in the dependent variable (Y) explained by the independent variable (X).

R² = 1 means perfect fit

R² = 0 means the model explains none of the variation
A higher R² indicates a better fit.



Q8: What is Multiple Linear Regression?

A8: Multiple Linear Regression is an extension of Simple Linear Regression where two or more independent variables (X₁, X₂, ..., Xn) are used to predict a dependent variable (Y). The equation looks like:
Y = b₀ + b₁X₁ + b₂X₂ + ... + bnXn



Q9: What is the main difference between Simple and Multiple Linear Regression?

A9: The main difference is the number of independent variables:

Simple Linear Regression has one independent variable.

Multiple Linear Regression has two or more independent variables.



Q10: What are the key assumptions of Multiple Linear Regression?

A10: The key assumptions are:

Linearity: The relationship between independent variables and the dependent variable is linear.

Independence: Observations are independent.

Homoscedasticity: Constant variance of residuals.

Normality: Residuals are normally distributed.

No multicollinearity: Independent variables should not be highly correlated with each other.



Q11: What is heteroscedasticity, and how does it affect MLR results?

Definition: Heteroscedasticity occurs when residuals have non-constant variance across predicted values.

Impact:

Biased standard errors → Incorrect significance tests (p-values).

Inefficient coefficient estimates (still unbiased but less precise).

Solution: Use robust standard errors (Huber-White) or transform the dependent variable (e.g., log transform).

Q12: How to improve an MLR model with high multicollinearity?

Remove highly correlated predictors (check correlation matrix).

Combine variables (e.g., PCA or factor analysis).

Regularization: Ridge/Lasso regression to penalize large coefficients.

Increase sample size (if possible).

VIF Check: Drop variables with VIF > 5–10.

Q13: Techniques for transforming categorical variables in regression?

Binary: Use dummy coding (0/1).

Multiclass:

One-Hot Encoding (k-1 dummies to avoid multicollinearity).

Effect Coding (compare levels to grand mean).

Ordinal: Assign scores (e.g., 1, 2, 3) or use polynomial contrasts.

Q14: Role of interaction terms in MLR?

Purpose: Model how the effect of one predictor depends on another.

Example: Sales ~ AdBudget * SocialMedia (effect of ads depends on social media presence).

Interpretation: The coefficient represents the additional effect when both predictors interact.

Q15: Interpretation of intercept in Simple vs. Multiple Regression?

Simple LR: Expected value of Y when X=0.

Multiple LR: Expected value of Y when all predictors = 0 (often meaningless if 0 is unrealistic).

Solution: Center predictors (subtract mean) for meaningful interpretation.

Q16: Significance of slope in regression?

Meaning: Change in Y per 1-unit increase in X, holding other predictors constant.

Effect on predictions: Larger slopes = stronger influence on Y (if significant).

Q17: How intercept provides context?

Baseline value: Sets the starting point for predictions.

Example: If intercept = 5, when all X=0, Y is expected to be 5 (context depends on variable scaling).



Q18: Limitations of R²?

Increases with more predictors (even useless ones) → Adjusted R² is better.

Doesn’t indicate causality.

Sensitive to outliers.

Useless for non-linear models.

Q19: Interpretation of large standard error for a coefficient?

Meaning: High uncertainty in the estimate → coefficient may not be reliable.

Causes:

Small sample size.

High multicollinearity.

High variability in data.

Q20: Identifying heteroscedasticity in residual plots?

Visual Check: Fan-shaped pattern in residuals vs. fitted plot.

Why Fix It?: Ensures valid hypothesis tests and confidence intervals.

Q21: High R² but low adjusted R²?
Meaning: Model includes too many unhelpful predictors.

Implication: Adjusted R² penalizes excess variables → suggests some predictors are irrelevant.

Assumptions matter (check residuals, multicollinearity, heteroscedasticity).

R² isn’t everything (use adjusted R², AIC, or cross-validation).

Interactions & categorical encoding improve model flexibility.

Q22: Why scale variables in Multiple Linear Regression?

Purpose: Ensure all predictors are on a comparable scale (e.g., for regularization, gradient descent, or interpretation).

Methods:

Standardization (Z-score: (X - mean)/std).

Normalization (Min-Max: scales to [0, 1]).

Impact: Prevents dominance by large-scale variables (e.g., income vs. age) and improves numerical stability.

Q27: Can Polynomial Regression Handle Multiple Variables?

Yes! Extend to multivariable polynomial regression and it's often referred to as multivariate polynomial regression when dealing with several independent variables. It extends the basic linear regression model by incorporating polynomial terms of the independent variables, allowing for more complex, non-linear relationships to be modeled.

Polynomial terms:
Instead of just using the independent variables directly, polynomial regression creates new features by raising the original variables to different powers (e.g., x², x³, x1*x2).

Multiple variables:
You can have multiple independent variables (e.g., x1, x2, x3) and create polynomial terms for each of them.

Interaction terms:
Polynomial regression can also model interactions between the variables, meaning the relationship between the variables can depend on each other. For example, a term like x1*x2 would indicate a potential interaction between variable x1 and x2.

Increased flexibility:
By including polynomial terms, the model can fit curves and surfaces more effectively, capturing complex relationships that linear regression alone might miss.

Example:
Imagine you want to predict house prices (the dependent variable) based on factors like size (x1) and number of bedrooms (x2). A simple linear regression might model a linear relationship. However, with polynomial regression, you could include terms like size², bedrooms², and even size\*bedrooms to capture more complex relationships. For instance, house prices might increase at a slower rate as the size gets very large, or the effect of bedrooms on price might depend on the size of the house.

In essence, polynomial regression is an extension of multiple linear regression that allows for non-linear relationships between the independent variables and the dependent variable, including interactions between the variables.



Q28: Limitations of Polynomial Regression

Overfitting: High-degree polynomials fit noise (use cross-validation).

Extrapolation Poor: Predictions outside the data range are unreliable.

Computational Cost: High degrees slow down training.

Q29: Methods to Evaluate Polynomial Degree

Cross-Validation: Compare MSE for different degrees.

Adjusted R²: Balances fit and complexity.

AIC/BIC: Penalize excessive parameters.
